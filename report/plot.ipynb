```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting x1,y1: PEFT w/ KLD, x2,y2: PBFT w/ CE
x1 = [0.05054, 0.01083, 0.01083, 0.00361]
y1 = [0.4304, 0.222, 0.0718, 0]
x2 = [0.5126, 0.556, 0.5993, 0.6173]
y2=[0.563, 0.3816, 0.695, 0.3724]

# Generate the plot
plt.figure(figsize=(10, 8))
plt.scatter(x1, y1, color='green', label='PEFT w/ KLD', s=150, marker='x')
plt.scatter(x2, y2, color='red', label='PBFT w/ CE', s=150, marker='x')

# diagonal line 
plt.plot([0.0, 0.9], [0.0, 0.9], 'k--', alpha=0.5)

# axis lable 
plt.xlabel('in-domain accuracy', fontsize=26)
plt.ylabel('out-of-domain accuracy', fontsize=26)
plt.title('RTE', fontsize=30)


# add grid
plt.grid(True, linestyle=':', alpha=0.7)

# add legend
plt.legend(fontsize=20)

# show the plot
plt.tight_layout()
plt.show()
```
